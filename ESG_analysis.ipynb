```python
ESG.columns = [col.split(' [')[0].strip() for col in ESG.columns] 
ESG = ESG.dropna(subset=['Country Name', 'Country Code', 'Series Name'])

numeric_cols = ESG.select_dtypes(include=[np.number]).columns
non_numeric_cols = ESG.select_dtypes(include=['object']).columns
ESG[non_numeric_cols] = ESG[non_numeric_cols].fillna('Unknown')

ESG = ESG.drop_duplicates()

year_columns = [col for col in ESG.columns if col.isdigit()] 
ESG[year_columns] = ESG[year_columns].apply(pd.to_numeric, errors='coerce')
ESG[year_columns] = ESG[year_columns].fillna(0)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


esg_long = pd.melt(ESG, 
                   id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'],
                   var_name='Year', 
                   value_name='Value')

esg_long['Year'] = pd.to_numeric(esg_long['Year'], errors='coerce') 
esg_long = esg_long.dropna(subset=['Year']) 
esg_long['Year'] = esg_long['Year'].astype(int) 

country = "United States"
indicator = "CO2 emissions (metric tons per capita)"  

us_data = esg_long[(esg_long['Country Name'] == country) & 
                   (esg_long['Series Name'] == indicator)]

plt.figure(figsize=(10,8))
plt.plot(us_data['Year'], us_data['Value'], marker='o', linestyle='-', color='blue')


plt.title(f"{indicator} in {country} (2022-2023)")
plt.xlabel("Year")
plt.ylabel("Value")
plt.xticks(rotation=45) 
plt.grid()
plt.show()
```

###
The graph illustrates CO2 emissions (metric tons per capita) in the United States from 1960 to 2023, highlighting irregularities and trends over time. Between 1960 and 1990, emissions remain at zero, likely due to missing or unreported data. A sharp spike occurs in the 1990s, suggesting possible anomalies or inconsistencies in recording. From 2022 to 2023, emissions start at approximately 19-20 metric tons per capita but show a significant downward trend, dropping to around 15 metric tons and eventually nearing zero. This decline could reflect changes in environmental policies, energy consumption, or data limitations. Overall, the graph underscores the need for consistent, accurate reporting to analyze CO2 emission trends effectively.

```python
country = "United States"
indicators = ["CO2 emissions (metric tons per capita)",
    "Renewable energy consumption (% of total final energy consumption)"
]

filtered_data = esg_long[(esg_long['Country Name'] == country) &
                         (esg_long['Series Name'].isin(indicators))]


plt.figure(figsize=(15,10))
for ind in indicators:
    indicator_data = filtered_data[filtered_data['Series Name'] == ind]
    plt.plot(indicator_data['Year'], indicator_data['Value'], marker='o', linestyle='-', label=ind)


plt.title(f"CO2 Emissions vs Renewable Energy Consumption in {country} (2022-2023)")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

### 
The graph illustrates the relationship between COâ‚‚ emissions per capita and renewable energy consumption in the United States over time. COâ‚‚ emissions showed a significant rise beginning in the late 1980s, peaking in the early 2000s, before experiencing a steady decline in the following years. In contrast, renewable energy consumption remained stagnant until the late 1990s but then began a steady upward trend. This inverse relationship suggests that the increase in renewable energy adoption has contributed to the decline in COâ‚‚ emissions, highlighting the positive impact of transitioning to cleaner energy sources. The data underscores the effectiveness of renewable energy policies and technological advancements in reducing carbon emissions, aligning with broader environmental, social, and governance (ESG) objectives. However, the abrupt changes at the start and end of the graph may indicate data inconsistencies or reporting gaps that warrant further investigation.
